```python
%matplotlib inline
from dolfin import *
from math import *
import numpy as np
import matplotlib.pyplot as plt

mresolution = 20
mesh = UnitIntervalMesh(mresolution)

p_order = 1;
V = FunctionSpace(mesh, "CG", p_order)
x = V.dofmap().tabulate_all_coordinates(mesh)
indices = np.argsort(x)

Eu = Function(V)
Eua = Eu.vector().array()
for j in indices:
   val = 1 + sin(10*x[j])
   Eua[j] = val
    
plt.plot(x[indices],Eua[indices],'-s');
```
